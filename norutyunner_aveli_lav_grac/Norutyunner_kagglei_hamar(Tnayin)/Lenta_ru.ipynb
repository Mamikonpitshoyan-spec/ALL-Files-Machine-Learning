```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7"
}
categories = {
    0: "https://lenta.ru/rubrics/russia/",
    1: "https://lenta.ru/rubrics/economics/",
    2: "https://lenta.ru/rubrics/forces/",
    3: "https://lenta.ru/rubrics/ussr/",
    4: "https://lenta.ru/rubrics/sport/",
    5: "https://lenta.ru/rubrics/health/",
    6: "https://lenta.ru/rubrics/realty/",
    7: "https://lenta.ru/rubrics/travel/",
    8: "https://lenta.ru/rubrics/science/"
}

url = "https://lenta.ru/rubrics/russia/"

r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, "html.parser")

links = soup.select("a[href*='/news/']")

article_links = []
for a in links:
    href = a.get("href")
    if href and re.search(r"/news/\d{4}/\d{2}/\d{2}/", href):

        if href.startswith("/"):
            href = "https://lenta.ru" + href

        article_links.append(href)


article_links = list(set(article_links))
print("Статей найдено:", len(article_links))


def get_article_text(url):
    r = requests.get(url, headers=headers, timeout=5)

    soup = BeautifulSoup(r.text, "html.parser")

    paragraphs = soup.select("div[data-type='text'] p")

    if not paragraphs:
        paragraphs = soup.select("p")

    text = " ".join(p.get_text(" ", strip=True) for p in paragraphs)
    print("LEN:", len(text))
    return text
    
    





data = []
MAX_ARTICLES_PER_CATEGORY = 500

for article_url in article_links[:MAX_ARTICLES_PER_CATEGORY]:
    try:
        text = get_article_text(article_url)
        if len(text) > 50:
            data.append({
                "category": "politics",
                "text": text
            })
        time.sleep(0.5)
    except Exception as e:
        print("Ошибка:", e)

df = pd.DataFrame(data)
print(df.shape)
df.head()
```

```python
all_data = []

for category, base_url in categories.items():
    print(f"\nПарсим категорию: {category}")

    article_links = set()   
    for page in range(1, 11):  
        page_url = base_url if page == 1 else f"{base_url}page/{page}/"
        print("Страница:", page_url)

        r = requests.get(page_url, headers=headers)
        soup = BeautifulSoup(r.text, "html.parser")

        links = soup.select("a[href*='/news/']")

        for a in links:
            href = a.get("href")

            if href and re.search(r"/news/\d{4}/\d{2}/\d{2}/", href):
                if href.startswith("/"):
                    href = "https://lenta.ru" + href
                article_links.add(href)

        time.sleep(0.5)

    print("Уникальных статей:", len(article_links))

    for article_url in article_links:
        try:
            text = get_article_text(article_url)

            if len(text) > 50:
                all_data.append({
                    "category": category,
                    "text": text
                })

            time.sleep(0.3)

        except Exception as e:
            print("Ошибка:", e)
```

```python
df = pd.DataFrame(all_data)
print(df.shape)

df.to_csv("train.csv", index=False)
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"\bфото\b.*?(?=\.|,|$)", "", text)
    text = re.sub(r"реклама", "", text)
    text=re.sub(r"[^a-zа-яё\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["text"] = df["text"].apply(clean_text)
df = df.drop_duplicates(subset=["text"]).reset_index(drop=True)
df.to_csv(df.to_csv("lenta_news_dataset_clean.csv", index=False)
, index=False)
```
