```python
from ultralytics import YOLO

models = {
    "aircraft": YOLO("C:/Users/USER/runs/detect/train/weights/best.pt"),
    "traffic":YOLO("C:/Users/USER/runs/detect/traffic2/weights/best.pt"),
    "drone": YOLO("C:/Users/USER/runs/detect/drone2/weights/best.pt"),
    "human": YOLO("C:/Users/USER/runs/detect/human/weights/best.pt"),
    "obstacle": YOLO("C:/Users/USER/runs/detect/tv_towers2/weights/best.pt"),
}

image_path = "OIP.jpg"

all_detections = []

for class_name, model in models.items():
    result = model(image_path, conf=0.25)[0]

    for box in result.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        conf = float(box.conf[0])

        all_detections.append({
            "class": class_name,
            "box": [x1, y1, x2, y2],
            "conf": conf
        })

print(all_detections)
```

```python
from ultralytics import YOLO
import cv2

models = {
    "aircraft": YOLO("C:/Users/USER/runs/detect/train/weights/best.pt"),
    "traffic": YOLO("C:/Users/USER/runs/detect/traffic2/weights/best.pt"),
    "drone": YOLO("C:/Users/USER/runs/detect/drone2/weights/best.pt"),
    "human": YOLO("C:/Users/USER/runs/detect/human/weights/best.pt"),
    "obstacle": YOLO("C:/Users/USER/runs/detect/tv_towers2/weights/best.pt"),
}

image_path = "images (1).jpg"

img = cv2.imread(image_path)

for class_name, model in models.items():
    result = model(image_path, conf=0.25)[0]

    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
        conf = float(box.conf[0])

        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 255, 255), 2)

        cv2.putText(
            img,
            f"{class_name} {conf:.2f}",
            (x1, max(20, y1 - 5)),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.6,
            (255, 255, 255),
            2
        )

output_path ="traffic_1.jpg"
cv2.imwrite(output_path, img)

print(f"Готовое фото сохранено: {output_path}")
```
