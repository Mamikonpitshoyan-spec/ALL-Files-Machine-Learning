```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = mpimg.imread('SirunStepan.jpg')
```

```python
plt.imshow(image)
```

```python
a, b, c = image.shape
```

```python
image = image.reshape(a * b, c)
```

```python
image.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

```python
label = model.fit_predict(image)
```

```python
label
```

```python
rgb_codes = model.cluster_centers_.astype(int)
```

```python
rgb_codes[0] = [247, 12, 173]
```

```python
image = rgb_codes[label]
```

```python
image = image.reshape(a, b, c)
```

```python
plt.imshow(image)
```
