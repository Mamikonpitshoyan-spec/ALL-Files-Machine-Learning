```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train
```

```python
df_test
```

```python
df_train.info()
```

```python
df_train = df_train.drop('id', axis=1)
df_test = df_test.drop('id', axis=1)
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train['Time_spent_Alone'] = df_train['Time_spent_Alone'].fillna(df_train['Time_spent_Alone'].median())
df_test['Time_spent_Alone'] = df_test['Time_spent_Alone'].fillna(df_test['Time_spent_Alone'].median())
```

```python
df_train['Stage_fear'].value_counts()
```

```python
df_train['Stage_fear'] = df_train['Stage_fear'].apply(lambda x: np.random.choice(a=['Yes', 'No'], p=[0.3, 0.7]) if pd.isna(x) else x)
df_test['Stage_fear'] = df_test['Stage_fear'].apply(lambda x: np.random.choice(a=['Yes', 'No'], p=[0.3, 0.7]) if pd.isna(x) else x)
```

```python
df_train['Social_event_attendance'] = df_train['Social_event_attendance'].fillna(df_train['Social_event_attendance'].median())
df_test['Social_event_attendance'] = df_test['Social_event_attendance'].fillna(df_test['Social_event_attendance'].median())
```

```python
df_train['Going_outside'] = df_train['Going_outside'].fillna(df_train['Going_outside'].median())
df_test['Going_outside'] = df_test['Going_outside'].fillna(df_test['Going_outside'].median())
```

```python
df_train['Drained_after_socializing'].value_counts()
```

```python
df_train['Drained_after_socializing'] = df_train['Drained_after_socializing'].apply(lambda x: np.random.choice(a=['No', 'Yes'], p=[0.6, 0.4]) if pd.isna(x) else x)
df_test['Drained_after_socializing'] = df_test['Drained_after_socializing'].apply(lambda x: np.random.choice(a=['No', 'Yes'], p=[0.6, 0.4]) if pd.isna(x) else x)
```

```python
df_train['Friends_circle_size'] = df_train['Friends_circle_size'].fillna(df_train['Friends_circle_size'].median())
df_test['Friends_circle_size'] = df_test['Friends_circle_size'].fillna(df_test['Friends_circle_size'].median())
```

```python
df_train['Post_frequency'] = df_train['Post_frequency'].fillna(3.0)
df_test['Post_frequency'] = df_test['Post_frequency'].fillna(3.0)
```

```python
df_train
```

```python
df_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in df_test.columns:
    if df_train[i].dtype == object:
        label = LabelEncoder()
        df_train[i] = label.fit_transform(df_train[i])
        df_test[i] = label.transform(df_test[i])
```

```python
X_train = df_train.drop('Personality', axis=1)
y_train = df_train['Personality']
X_test = df_test.copy()
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.pipeline import Pipeline
```

```python
estimator = DecisionTreeClassifier(max_depth=6, splitter='best', criterion='entropy')
```

```python
pca = PCA()
model = AdaBoostClassifier(estimator)
pipe = Pipeline([('pca', pca), ('model', model)])
```

```python
params = {
    'pca__n_components':[1, 2, 3, 4, 5, 6, 7, 8],
    'model__n_estimators':[80, 100, 120, 150],
    'model__learning_rate':[0.01, 0.1, 1, 10]
}
```

```python
final_model = GridSearchCV(pipe, params, cv=3, verbose=2)
```

```python
final_model.fit(X_train, y_train)
```

```python
final_model.best_params_
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['Personality'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index=False, index_label=False)
```
