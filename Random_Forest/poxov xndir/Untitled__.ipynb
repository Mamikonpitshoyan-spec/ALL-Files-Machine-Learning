```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
test_ids = df_test['event_id'].copy()
```

```python
drop_cols = ['event', 'time_to_hit_hours', 'event_id']

X_train = df_train.drop(columns=drop_cols)
y_train = df_train['event']

X_test = df_test.drop(columns=drop_cols, errors='ignore')
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
param_grid = {
    'n_estimators': [400],
    'min_samples_leaf': [2, 3],
    'min_samples_split': [4, 6],
    'max_depth': [4, 8, 10, 14]
}
```

```python
assert list(X_train.columns) == list(X_test.columns)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=400,
    min_samples_split=5,
    min_samples_leaf=2,
    n_jobs=-1,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
final_model = GridSearchCV(model,param_grid , cv=3, verbose=2)
```

```python
final_model.fit(X_train, y_train)
```

```python
base_prob = rf.predict_proba(X_test)[:, 1]
```

```python
import numpy as np

prob_12h = np.clip(base_prob * 0.6, 0, 1)
prob_24h = np.clip(base_prob * 0.8, 0, 1)
prob_48h = np.clip(base_prob * 1.0, 0, 1)
prob_72h = np.clip(base_prob * 1.2, 0, 1)
```

```python
final_model.best_params_
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub = pd.DataFrame({
    'event_id': test_ids,
    'prob_12h': prob_12h,
    'prob_24h': prob_24h,
    'prob_48h': prob_48h,
    'prob_72h': prob_72h
})

df_sub.to_csv('sub_IM_UXXELUC_HETO2.csv', index=False,index_label=False)
```
