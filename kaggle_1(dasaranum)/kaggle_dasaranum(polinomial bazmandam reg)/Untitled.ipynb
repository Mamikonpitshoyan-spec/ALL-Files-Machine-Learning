```python
X_train = df_train.drop('accident_risk', axis=1)
y_train = df_train['accident_risk']
X_test = df_test.copy()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
poly = PolynomialFeatures()
model = LinearRegression()
```

```python
pipe = Pipeline([('poly', poly), ('model', model)])
```

```python
params = {
    'poly__degree':[1, 2, 3],
    'poly__include_bias':[False]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
final_model = GridSearchCV(pipe, params, cv=3, verbose=2)
```

```python
final_model.fit(X_train, y_train)
```

```python
final_model.best_score_
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['accident_risk'] = y_pred
```
