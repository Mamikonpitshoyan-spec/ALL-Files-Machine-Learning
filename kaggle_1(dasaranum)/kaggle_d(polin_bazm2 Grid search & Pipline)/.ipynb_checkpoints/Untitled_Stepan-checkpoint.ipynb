```python
X_train = df_train.drop('Premium Amount', axis=1)
y_train = df_train['Premium Amount']
X_test = df_test.copy()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
poly = PolynomialFeatures()
model = LinearRegression()
```

```python
pipe = Pipeline([('X', poly), ('y', model)])
```

```python
params = {
    'X__degree':np.arange(1, 5),
    'X__include_bias':[False]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
finalModel = GridSearchCV(pipe, params, cv=2, verbose=2)
```

```python
finalModel.fit(X_train, y_train)
```

```python
finalModel.best_score_
```

```python
finalModel.best_params_
```

```python
y_pred = finalModel.predict(X_test)
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['Premium Amount'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index_label=False, index=False)
```
