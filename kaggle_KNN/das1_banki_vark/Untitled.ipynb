```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train
```

```python
df_test
```

```python
df_train = df_train.drop('id', axis=1)
df_test = df_test.drop('id', axis=1)
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
set(df_test['person_home_ownership']).difference(set(df_train['person_home_ownership']))
```

```python
df_train['person_home_ownership'] = label.fit_transform(df_train['person_home_ownership'])
df_test['person_home_ownership'] = label.fit_transform(df_test['person_home_ownership'])
```

```python
set(df_test['loan_intent']).difference(set(df_train['loan_intent']))
```

```python
df_train['loan_intent'] = label.fit_transform(df_train['loan_intent'])
df_test['loan_intent'] = label.fit_transform(df_test['loan_intent'])
```

```python
set(df_test['loan_grade']).difference(set(df_train['loan_grade']))
```

```python
df_train['loan_grade'] = label.fit_transform(df_train['loan_grade'])
df_test['loan_grade'] = label.fit_transform(df_test['loan_grade'])
```

```python
df_train.info()
```

```python
set(df_test['cb_person_default_on_file']).difference(set(df_train['cb_person_default_on_file']))
```

```python
df_train['cb_person_default_on_file'] = label.fit_transform(df_train['cb_person_default_on_file'])
df_test['cb_person_default_on_file'] = label.fit_transform(df_test['cb_person_default_on_file'])
```

```python
X_train = df_train.drop('loan_status', axis=1)
y_train = df_train['loan_status']
X_test = df_test.copy()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['loan_status'] = y_pred
```

```python
df_sub.to_csv('sub.csv',index = False,index_label = False)
```
