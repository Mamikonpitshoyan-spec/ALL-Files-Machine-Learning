```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train
```

```python
df_test
```

```python
df_train = df_train.drop('id', axis=1)
df_test = df_test.drop('id', axis=1)
```

```python
df_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
df_train['gender'] = label.fit_transform(df_train['gender'])
df_test['gender'] = label.fit_transform(df_test['gender'])
```

```python
df_train['ethnicity'] = label.fit_transform(df_train['ethnicity'])
df_test['ethnicity'] = label.fit_transform(df_test['ethnicity'])
```

```python
df_train['education_level'] = label.fit_transform(df_train['education_level'])
df_test['education_level'] = label.fit_transform(df_test['education_level'])
```

```python
df_train['income_level'] = label.fit_transform(df_train['income_level'])
df_test['income_level'] = label.fit_transform(df_test['income_level'])
```

```python
df_train['smoking_status'] = label.fit_transform(df_train['smoking_status'])
df_test['smoking_status'] = label.fit_transform(df_test['smoking_status'])
```

```python
df_train['employment_status'] = label.fit_transform(df_train['employment_status'])
df_test['employment_status'] = label.fit_transform(df_test['employment_status'])
```

```python
X_train = df_train.drop('diagnosed_diabetes', axis=1)
y_train = df_train['diagnosed_diabetes']
X_test = df_test.copy()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
params = {
    'criterion':["gini", "entropy", "log_loss"],
    'splitter': ['best', 'random'],
    'max_depth':[4,5,6]
}
```

```python
final_model = GridSearchCV(model,params,cv = 3,verbose = 2)
```

```python
final_model.fit(X_train,y_train)
```

```python
final_tree = final_model.best_estimator_
```

```python
final_model.best_params_
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(10, 6), dpi=300)
plot_tree(final_tree, feature_names=X_train.columns)
plt.show()
```

```python
y_pred = final_model.predict(X_test)
```
