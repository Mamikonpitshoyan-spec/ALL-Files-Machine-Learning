```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import cv2
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python

labels = {
    'Full  Water level': 0,
    'Half water level': 1,
    'Overflowing': 2
}

datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    channel_shift_range=20,
    fill_mode='nearest'
)

X = []
y = []

AUG_COPIES = 3

for class_name in os.listdir('.'):
    if class_name not in labels:
        continue
    class_path = class_name
    try:
        for img_file in os.listdir(class_path):
            img_path = os.path.join(class_path, img_file)
            image = cv2.imread(img_path)

            if image is None:
                continue

            image = cv2.resize(image, (200, 200))

            if image.shape[2] == 4:
                image = cv2.cvtColor(image, cv2.COLOR_BGRA2BGR)

            image = image / 255.0

            X.append(image)
            y.append(labels[class_name])

            image_exp = np.expand_dims(image, 0)
            aug_iter = datagen.flow(image_exp, batch_size=1)
            for _ in range(AUG_COPIES):
                aug_image = next(aug_iter)[0]
                X.append(aug_image)
                y.append(labels[class_name])

    except Exception as e:
        print(f"Ошибка при обработке {class_name}/{img_file}: {e}")
        continue

X = np.array(X)
y = np.array(y)

print(f"Итоговое количество изображений: {X.shape[0]}")
print(f"Итоговое количество меток: {y.shape[0]}")
```

```python
len(X)
```

```python
len(y)
```

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y, num_classes=3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
X_train.shape
```

```python
from keras import Sequential
from keras.layers import Dense, MaxPooling2D, Conv2D, Flatten, Dropout, BatchNormalization
from keras.optimizers import Adam, RMSprop
```

```python
model = Sequential()
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', input_shape=(200, 200, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(units=120, activation='relu'))
model.add(Dropout(0.5))
model.add(BatchNormalization())
model.add(Dense(units=100, activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Dense(units=60, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=3, activation='softmax'))
```

```python
model.compile(optimizer=RMSprop(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10)
```

```python
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
```
