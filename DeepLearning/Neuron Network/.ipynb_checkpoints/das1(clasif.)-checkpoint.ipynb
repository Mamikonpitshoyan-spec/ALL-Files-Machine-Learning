```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df_train['Heart Disease'] = label.fit_transform(df_train['Heart Disease'])
```

```python
X_train = df_train.drop('Heart Disease', axis=1)
y_train = df_train['Heart Disease']
X_test = df_test.copy()
```

```python
X_train = X_train.drop('id', axis=1)
X_test = X_test.drop('id', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, num_classes=2)
```

```python
y_train.shape
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(13, )))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_split=0.30, epochs=5)
```

```python
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = y_pred[::, 1]
```

```python
y_pred
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['Heart Disease'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index=False, index_label=False)
```
