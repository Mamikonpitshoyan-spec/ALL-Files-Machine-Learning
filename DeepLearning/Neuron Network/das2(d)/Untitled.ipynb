```python
X_train = df_train.drop('yield', axis=1)
y_train = df_train['yield']
X_test = df_test.copy()
```

```python
X_train = X_train.drop('id', axis=1)
X_test = X_test.drop('id', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(units = 96,activation='relu',input_shape = (17, )))
model.add(Dense(units = 64,activation='relu'))
model.add(Dense(units = 36,activation='relu'))
model.add(Dense(units = 18,activation='relu'))
model.add(Dense(units = 9,activation='relu'))
model.add(Dense(units = 1,activation='linear'))
```

```python
model.compile(optimizer = 'adam' , loss = 'mae')
```

```python
model.fit(X_train, y_train, validation_split=0.30, epochs=35)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['yield'] = y_pred
```

```python
df_sub.to_csv('sub6.csv',index = False, index_label= False)
```
