```python
import os
import shutil

base_path = 'Testing/Testing/'

smoking_dir = os.path.join(base_path, 'smoking')
notsmoking_dir = os.path.join(base_path, 'not_smoking')

os.makedirs(smoking_dir, exist_ok=True)
os.makedirs(notsmoking_dir, exist_ok=True)

for file in os.listdir(base_path):

    file_path = os.path.join(base_path, file)

    if os.path.isdir(file_path):
        continue

    filename_lower = file.lower()

    if filename_lower.startswith('smoking_'):
        shutil.move(file_path, os.path.join(smoking_dir, file))

    elif filename_lower.startswith('notsmoking_'):
        shutil.move(file_path, os.path.join(notsmoking_dir, file))

print("Готово ✅")
```

```python
X_train = []
y_train = []
base_path = 'Training/Training'
data = {'not_smoking':0, 'smoking':1}
for i in os.listdir(base_path):
    if i.startswith('.'):
        continue
    for j in os.listdir(f'{base_path}/{i}/'):
        try:
            image = mpimg.imread(f'{base_path}/{i}/{j}')
            image = cv2.resize(image, (250,250))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            image = image.reshape(250,250,1)
            X_train.append(image)
            y_train.append(data[i])
        except:
             continue
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_test = []
y_test = []
base_path = 'Validation/Validation/'
data = {'not_smoking':0, 'smoking':1}
for i in os.listdir(base_path):
    if i.startswith('.'):
        continue
    for j in os.listdir(f'{base_path}/{i}/'):
        try:
            image = mpimg.imread(f'{base_path}/{i}/{j}')
            
            image = cv2.resize(image, (250,250))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            image = image.reshape(250,250,1)
            X_test.append(image)
            y_test.append(data[i])
        except:
             continue
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train,num_classes=2)
y_test = to_categorical(y_test,num_classes=2)
```

```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(250, 250, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=16, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer = Adam(learning_rate=0.001),loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs = 20, batch_size = 32)
```

```python
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc * 100:.2f}%")
```
