```python
data = {
    'cat':0,
    'dog':1
}
```

```python
X_train = []
y_train = []
for i in os.listdir('animals/'):
    for j in os.listdir(f'animals/{i}/'):
        image = mpimg.imread(f'animals/{i}/{j}')
        image = cv2.resize(image, (200, 200))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = image.reshape(200, 200, 1)
        X_train.append(image)
        y_train.append(data[i])
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.utils import shuffle

X_train, y_train = shuffle(X_train, y_train, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, num_classes=2)
```

```python
y_train.shape
```

```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', input_shape=(200, 200, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_split=0.15, epochs=10)
```

```python
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
```
