```python
X_train = []
y_train = []
data = {
   'angry':0, 'disgusted':1, 'fearful':2, 'happy':3, 'neutral':4, 'sad':5, 'surprised':6 
}
for i in os.listdir('train/'):
    for j in os.listdir(f'train/{i}/'):
        try:
            image = mpimg.imread(f'train/{i}/{j}')
            image = image.reshape(48, 48, 1)
            X_train.append(image)
            y_train.append(data[i])
        except:
            print(j)
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_test = []
y_test = []
data = {
   'angry':0, 'disgusted':1, 'fearful':2, 'happy':3, 'neutral':4, 'sad':5, 'surprised':6 
}
for i in os.listdir('test/'):
    for j in os.listdir(f'test/{i}/'):
        try:
            image = mpimg.imread(f'test/{i}/{j}')
            image = image.reshape(48, 48, 1)
            X_test.append(image)
            y_test.append(data[i])
        except:
            print(j)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, num_classes=7)
y_test = to_categorical(y_test, num_classes=7)
```

```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(48, 48, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(units=7, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=32)
```

```python
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
```
