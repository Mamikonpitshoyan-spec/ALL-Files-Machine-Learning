```python
params = {
    'X__degree':np.arange(1, 7),
    'X__include_bias':[False]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
finalModel = GridSearchCV(pipe, params, cv=2, verbose=2)
```
