```python
df_train = df_train.drop('id', axis=1)
df_test = df_test.drop('id', axis=1)
```

```python
df_train = df_train.drop('Surname', axis=1)
df_test = df_test.drop('Surname', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
set(df_test['Geography']).difference(set(df_train['Geography']))
```

```python
df_train['Geography'] = label.fit_transform(df_train['Geography'])
df_test['Geography'] = label.transform(df_test['Geography'])
```

```python
df_train['Gender'] = label.fit_transform(df_train['Gender'])
df_test['Gender'] = label.transform(df_test['Gender'])
```

```python
y_train = df_train['Exited']
X_train = df_train.drop(columns=['Exited'])
X_test = df_test.copy()
```

```python
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
model = XGBClassifier(
    objective='binary:logistic',
    random_state=42,
    eval_metric='logloss'
)
pipe = Pipeline([('model', model)])
```

```python
params = {
    "model__n_estimators": [200, 400, 500],
    "model__learning_rate": [0.01, 0.05, 0.1],
    "model__max_depth": [3, 5, 8]
}
```

```python
final_model = GridSearchCV(pipe, params, cv=3, verbose=2, n_jobs=-1)
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['Exited'] = y_pred
```

```python
df_sub.to_csv('sub.csv',index=False,index_label=False)
```
